## Lasso Logistic Regression with 'RAZON' as Target

Since the `RAZON` column has been transformed into a binary target variable, I will now re-split the data into training and testing sets to reflect this change. Then, I will define the features (X) and the target (y) for both the training and testing datasets and set up a preprocessing pipeline for different types of features, followed by training and evaluating a Lasso Logistic Regression model.

````python

df_train = df[df['FIn'].dt.year.isin([2023, 2024])].copy()
df_test = df[df['FIn'].dt.year == 2025].copy()


display(df_train.head())
display(df_test.head())
````

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report

y_train = df_train['RAZON']
X_train = df_train.drop(columns=['RAZON', 'FIn', 'Fter'])

y_test = df_test['RAZON']
X_test = df_test.drop(columns=['RAZON', 'FIn', 'Fter'])

categorical_features = X_train.select_dtypes(include=['object', 'category']).columns
numerical_features = X_train.select_dtypes(include=['int64', 'float64']).columns


preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])


custom_class_weights = {'EXITOSO': 1, 'NO EXITOSO': 6}

lasso_logistic_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(penalty='l1', solver='liblinear', random_state=42, C=1.0, class_weight=custom_class_weights)) # C is the inverse of regularization strength
])


lasso_logistic_pipeline.fit(X_train, y_train)

print("Lasso Logistic Regression model trained successfully with custom weights.")


y_pred = lasso_logistic_pipeline.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"\nAccuracy: {accuracy:.4f}")
print(report)
````

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV, StratifiedKFold


y_train = df_train['RAZON']
X_train = df_train.drop(columns=['RAZON', 'FIn', 'Fter'])

y_test = df_test['RAZON']
X_test = df_test.drop(columns=['RAZON', 'FIn', 'Fter'])


categorical_features = X_train.select_dtypes(include=['object', 'category']).columns
numerical_features = X_train.select_dtypes(include=['int64', 'float64']).columns


preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])


custom_class_weights = {'EXITOSO': 1, 'NO EXITOSO': 6}


lasso_logistic_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(penalty='l1', solver='liblinear', random_state=42, C=1.0, class_weight=custom_class_weights)) # C is the inverse of regularization strength
])


param_grid = {'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100]}


cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)


grid_search = GridSearchCV(lasso_logistic_pipeline, param_grid, cv=cv, scoring='f1_macro', n_jobs=-1, verbose=1)


grid_search.fit(X_train, y_train)

print(f"Best C parameter: {grid_search.best_params_['classifier__C']}")
print(f"Best cross-validation f1_macro score: {grid_search.best_score_:.4f}")


best_lasso_model = grid_search.best_estimator_


y_pred_best = best_lasso_model.predict(X_test)
accuracy_best = accuracy_score(y_test, y_pred_best)
report_best = classification_report(y_test, y_pred_best)

print(f"\nAccuracy on the test set with best C (optimizing for f1_macro): {accuracy_best:.4f}")
print("\nClassification Report with best C (optimizing for f1_macro):\n", report_best)
````

````python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report


print("Training model with class_weight='balanced' (C=1.0) for demonstration...")

balanced_weight_pipeline_demo = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(penalty='l1', solver='liblinear', random_state=42, C=1.0, class_weight='balanced'))
])

balanced_weight_pipeline_demo.fit(X_train, y_train)

y_pred_balanced_demo = balanced_weight_pipeline_demo.predict(X_test)
accuracy_balanced_demo = accuracy_score(y_test, y_pred_balanced_demo)
report_balanced_demo = classification_report(y_test, y_pred_balanced_demo)

print(f"\nAccuracy on the test set with class_weight='balanced' (C=1.0): {accuracy_balanced_demo:.4f}")
print("\nClassification Report with class_weight='balanced' (C=1.0):\n", report_balanced_demo)
````

**Summary:**
*   The model with `class_weight='balanced'` achieves a higher overall accuracy (0.40 vs 0.35) and a better macro average F1-score (0.40 vs 0.31) compared to the GridSearchCV optimized model (which used custom weights and optimized for `f1_macro`).
*   The `class_weight='balanced'` model also shows a more balanced recall between the two classes (30% for 'EXITOSO' and 65% for 'NO EXITOSO'), while the GridSearchCV model still heavily favors the 'NO EXITOSO' class in terms of recall (99%), largely ignoring 'EXITOSO'.

````python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

print("\n--- Confusion Matrix for Best GridSearchCV Model (f1_macro optimized) ---")
y_pred_best = best_lasso_model.predict(X_test)
cm_best = confusion_matrix(y_test, y_pred_best, labels=best_lasso_model.classes_)
disp_best = ConfusionMatrixDisplay(confusion_matrix=cm_best, display_labels=best_lasso_model.classes_)
disp_best.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix: Best GridSearchCV Model')
plt.show()


print("\n--- Confusion Matrix for Model with class_weight='balanced' ---")
y_pred_balanced_demo = balanced_weight_pipeline_demo.predict(X_test)
cm_balanced = confusion_matrix(y_test, y_pred_balanced_demo, labels=balanced_weight_pipeline_demo.classes_)
disp_balanced = ConfusionMatrixDisplay(confusion_matrix=cm_balanced, display_labels=balanced_weight_pipeline_demo.classes_)
disp_balanced.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix: Model with class_weight=\'balanced\'')
plt.show()
````

This exercise explicitly shows that even with `class_weight='balanced'`, the classification report remains the same as the default model (C=1.0), reinforcing the observation that `class_weight` is not having the expected impact. This warrants further investigation or considering alternative methods to address class imbalance.
